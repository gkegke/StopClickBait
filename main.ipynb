```python
from youtube_transcript_api import YouTubeTranscriptApi
from urllib.parse import urlparse, parse_qs

import openai
```

```python
## Add api key
#openai.api_key = ""

video_url = "https://www.youtube.com/watch?v=4Bdc55j80l8"

model_id = "text-davinci-003"
```

```python

try: 
    url_data = urlparse(video_url)
    video_id = parse_qs(url_data.query)["v"][0]
except Exception as e:
    print("url {} doesn't seem to be valid".format(video_url))
    raise e

video_id
```

```python
transcript = YouTubeTranscriptApi.get_transcript(video_id)
```

```python
transcript = " ".join([part["text"] for part in transcript])

print(transcript)
```

```python
response = openai.Completion.create(
  model="text-davinci-003",
  prompt="Summarize: {}".format(transcript)
  temperature=0,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)
```
